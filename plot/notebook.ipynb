```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_data(state):
    file = f'../data/{state}.csv'
    data = np.loadtxt(file, delimiter=',', encoding='utf-8-sig')
    return data[:,0], data[:,1], data[:, 2]
```

```python
def plot_axis(axis, title, ax, data):
        ax.set_title(title)
        for i in range(len(data)):
            d, label = data[i]
            ax.boxplot(
                d[axis],
                positions=[i],
                labels=[label],
                showfliers=False,
            )
```

```python
def plot_data(data, title):
    axes = ['X', 'Y', 'Z']
    fig, ax = plt.subplots(len(axes), figsize=(5, 15))
    fig.set_dpi(100)
    fig.suptitle(title, fontsize=15)
    for i, axis in enumerate(axes):
        plot_axis(i, f'Ось {axis}', ax[i], data)
    plt.show()
    fig.savefig('result.png', bbox_inches='tight')
```

```python
def plot(states):
    data = [(load_data(state), caption) for state, caption in states]
    plot_data(data, 'Показания акселерометра в зависимости от положения устройства')
```

```python
states = [
    ('table', 'Устройство на столе'),
    ('hand', 'Устройство в руке'),
]
```

```python
plot(states)
```
